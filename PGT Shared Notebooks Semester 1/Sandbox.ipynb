```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters for the distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate 1000 points for the probability density function
x = np.linspace(-5, 5, 1000)  # Generating 1000 points between -5 and 5
pdf = norm.pdf(x, mu, sigma)  # Probability Density Function

# Plot the PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='PDF (Normal Distribution)')
plt.title('Probability Density Function')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the parameters for the distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate random samples from the distribution
num_samples = 1000
samples = np.random.normal(mu, sigma, num_samples)

# Create a histogram of the samples
plt.figure(figsize=(8, 6))
plt.hist(samples, bins=30, density=True, alpha=0.7, label='Histogram')

# Plot the PDF for comparison
x = np.linspace(-5, 5, 1000)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, label='PDF (Normal Distribution)', color='red')

plt.title('Histogram and Probability Density Function')
plt.xlabel('Values')
plt.ylabel('Density/Probability')
plt.legend()
plt.grid(True)
plt.show()
```

Standard error : measures the sample mean (the mean of the samples itself) compared to standard deviation, which is the ariability of the datapoints in the complete population

```python
# Generating random data (example: normal distribution)
np.random.seed(42)  # Setting seed for reproducibility
population_mean = 50
population_std = 10
sample_size = 30
num_samples = 100

# Generate multiple samples from the population
samples = np.random.normal(population_mean, population_std, size=(num_samples, sample_size))

# Calculate the means of the samples
sample_means = np.mean(samples, axis=1)

# Calculate standard error
standard_error = np.std(sample_means, ddof=1) / np.sqrt(len(sample_means))

print("Sample Means:", sample_means)
print("Standard Error:", standard_error)
```
